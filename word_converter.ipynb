```python
def  bytes_to_float(aa):
    import struct
    lis=[aa[1],aa[0],aa[3],aa[2]]
    aa= bytearray(lis) 
    
    out=struct.unpack('>f', aa)[0]
    print (f"{out:0.22f}", end=" . ")



   
    return(out)
```
